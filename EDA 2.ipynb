# Isolation Forest

```python
import pandas as pd
from sklearn.ensemble import IsolationForest
```

```python
data=pd.read_csv("Downloads\\iris1.csv",index_col=0)
data_encoded=pd.get_dummies(data)
data_encoded
```

```python
clf=IsolationForest(random_state=10,contamination=0.01)
clf.fit(data_encoded)
```

```python
y_pred_outliers=clf.predict(data_encoded)
y_pred_outliers
```

```python
data_encoded['Outliers']=y_pred_outliers
data_encoded
```

```python
data_encoded[data_encoded.Outliers==-1]
```

```python
data_encoded.drop('Outliers',axis=1,inplace=True)
data_encoded
```

```python
data_encoded.loc[150]=[20,40,30,50,1,0,0]
data_encoded
```

```python
clf=IsolationForest(random_state=10,contamination=0.01)
clf.fit(data_encoded)
```

```python
y_pred_outliers=clf.predict(data_encoded)
y_pred_outliers
```

```python
data_encoded['Outliers']=y_pred_outliers
data_encoded
```

```python
data_encoded[data_encoded.Outliers==-1]
```
